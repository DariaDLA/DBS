```python
from pyexistdb import db
```

```python
urlexist = "http://localhost:8080/exist/"
```

```python
db = db.ExistDB(urlexist)
```

```python
# funktioniert nicht
xquery = '''
let $x:= doc("/db/employment/Uebung1.xml")/*
return $x/gemeinde/node()'''
```

```python
# auch nicht
xquery = '''
let $gemeinden := doc('/db/employment/Uebung1.xml')/*
for $x in $gemeinden/gemeinde
order by $x/women/SELF_EMPL_W div $x/total/EMPL_TOTAL descending
return (data($x/@LAU_NAME), $x/women/SELF_EMPL_W div $x/total/EMPL_TOTAL, $x/women/SELF_EMPL_W, $x/total/EMPL_TOTAL)'''
```

```python
# das funktioniert aber schon!
xquery = '''
let $x:= doc("/db/employment/Uebung1.xml")/*
return $x/*'''
```

```python
res = db.executeQuery(xquery)
```

```python
hits = db.getHits(res)
```

```python
hits
```

**Interessant: Aus irgendeinem Grund bekommt das Element "gemeinde" nach der Abfrage ein zusätzliches Attribut "xmlns"**

```python
for i in range(hits):
    print(str(db.retrieve(res,0)))
```
